Arithmetic Operations

```python
import numpy as np
arr1=np.array([[5,3,2],[2,1,3],[7.,4,6]])
arr2=np.array([[0,6,2],[6,3,4],[8,6,4]])
print(arr1+arr2)
print()
print(arr1-arr2)
print()
print(np.dot(arr1,arr2))
```

aggrgate functions 

```python
import numpy as np
arr1=np.array([[5,3,2],[2,1,3],[7.,4,6]])
arr2=np.array([[0,6,2],[6,3,4],[8,6,4]])
print(np.sum(arr1))
print(np.mean(arr1))
print(np.median(arr1))
print("max of arr1 is ",np.max(arr1))
print("max of arr2 is",np.max(arr2))
print("Min of arr1 is",int(np.min(arr1)))
print("Min of arr2 is",np.min(arr2))
print(np.std(arr2))
```

indexing and slicing in numpy

```python
import numpy as np
arr1=np.array([0,1,2,3,2,1,6,91,99928])
print(np.std(arr1))
print(arr1[::-1])
```
