matplot function

```python
import matplotlib.pyplot as plt
x=[32,45,12,11]
y=[1,23,66,41]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
a=[11,22,33,44,55]
b=[11,22,33,44,55]
plt.plot(a,b,color="red",linestyle="--")
plt.title("Linear graph")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

more than 1 lines in matplotlib

```python
import matplotlib.pyplot as plt
a=[11,23,43,11,111]
b=[224,33,221,45,12]
c=[12,21,100,22,2]
plt.plot(a,b,color="blue",linestyle="--",label="a")
plt.plot(b,c,color="red",linestyle="--",label="v")
plt.legend()
plt.show()
```
