Merging & Joining

```python
import pandas as pd
data1={"id" : [0,1],"name" :["Sundar","Kp ganesh"], "age" : [44,78]}
data2={"id" : [1,0], "Gender" : ["M","NB"]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
pd.merge(df1,df2, on="id")
```
