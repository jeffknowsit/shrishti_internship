reading csv

```python
import pandas as pd
data=pd.read_csv("Social_Network_Ads.csv")
data.head()
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

classifing and  splliting into X and Y

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

applying decision tres clasifier to the  data model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

training again (more accurate)

```python
clf2 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
clf2.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
print(clf2.predict(sc.transform([[51,87000]])))
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

descision view


```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    classifier,
    X_train,
    response_method="predict_proba",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=data.columns[0],
    ylabel=data.columns[1],
)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,s=20,edgecolors="k")
plt.title("Descision tree classifier")
plt.show()
```

full tree structure view

```python
from sklearn import tree
import matplotlib.pyplot as plt

def plot_decision_tree(clf_object,input_columns,output_columns):
    tree.plot_tree(
        clf_object,
        filled=True,
        feature_names=input_columns,
        class_names=output_columns,
        rounded=True,
    )
    plt.savefig("decision_tree_4k.png", dpi=200, bbox_inches="tight")
    plt.show()
plot_decision_tree(classifier,["age","salary"],["yes","no"])
```

```python
print("Age",classifier.feature_importances_[0])
print("Estimated Salary",classifier.feature_importances_[1])
```

```python
plot_decision_tree(classifier,["Age","Estimated Salary"],["Purchased","Not Purchased"])
```
