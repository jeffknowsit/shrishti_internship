Check whether a given number exists in a set.


```python
set_new={1,89,32,89,61,67,98,87,2,1,5}
n=int(input())
flag=0
for i in set_new:
    if(i==n):
        flag=1
        break;
if(flag==1):
    print(" found " ,n)
else:
    print("not found " ,n)
```

Check whether one set is a subset of another.


```python
print("elements in set 1")
set1=eval(input())
print()
print("elements in set 2")
print()
set2=eval(input())
print()
z=set1.intersection(set2)
if(len(z)==0):
    print("set1 and set2 are not subset")
    exit(1)
else:
    print("set1 and set2 are subset")
    
```

Check whether two sets are disjoint.


```python
print("elements in set 1")
set1=eval(input())
print()
print("elements in set 2")
print()
set2=eval(input())
print()
z=set1.intersection(set2)
if(len(z)==0):
    print("set1 and set2 are disjoint sets")
    exit(1)
else:
    print("set1 and set2 are not disjoint set")
    
```
