```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
x=dataset.iloc[:,[3,4]].values
dataset
```

elbow method to optimal number

```python
from pandas._libs import indexing
from sklearn.cluster import KMeans
wcms=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(x)
    wcms.append(kmeans.inertia_)
plt.plot(range(1,11),wcms)
plt.title("The elbow method")
plt.xlabel("Number of customers")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans== 1, 0], x[y_kmeans== 1, 1],s=100,c='blue',label="Cluster 3")
plt.scatter(x[y_kmeans== 2, 0],x[y_kmeans== 2, 1],s=100,c='magenta',label="Cluster 3")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='green',label="Cluster 4")
plt.legend()
plt.xlabel("Anuual Income")
plt.ylabel("Spending score (1-100)")
plt.show()
```
